Importing libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import math
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, mean_squared_error, cohen_kappa_score, matthews_corrcoef
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier

```

Reading dataset

```python
df = pd.read_csv('kaggle_2.csv')
print (df.status.value_counts())
df.head()
```

data pre-processing

```python
df = df.drop(['url'], axis='columns')
Label = LabelEncoder()
df['status'] = Label.fit_transform(df['status'])
print (df.status.value_counts())
df.head()
```

```python
input = df.drop (['status'], axis='columns')
target = df.status
print(target.head())
input.head()
```

Data splitting

```python
X_train, X_test, y_train, y_test = train_test_split(input, target, test_size=0.3)
print (len (X_test))
print (len (X_train))
```

Logistic Regression

```python
model = LogisticRegression ()
model.fit (X_train, y_train)
```

```python
y_pred = model.predict (X_test)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
auc_score = roc_auc_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

sensitivity = conf_matrix[1, 1] / (conf_matrix[1, 0] + conf_matrix[1, 1])
specificity = conf_matrix[0, 0] / (conf_matrix[0, 0] + conf_matrix[0, 1])

kappa = cohen_kappa_score(y_test, y_pred)
mcc = matthews_corrcoef(y_test, y_pred)
g_mean = (sensitivity * specificity)**0.5

print("Classification Report:\n", report)
print("Confusion Matrix:\n", conf_matrix)
print("AUC Score:", auc_score)
print("MSE:", mse)
print("G-mean:", g_mean)
print("Kappa:", kappa)
print("MCC:", mcc)
print("Confusion Matrix:\n", conf_matrix)
```
